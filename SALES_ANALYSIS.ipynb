```python
import pandas as pd
df=pd.read_csv(r"/content/sales_dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df["Region"]=df["Region"].fillna(df["Region"].mode()[0])
df
```

```python
df["Product"]=df["Product"].fillna(df["Product"].mode()[0])
df
```

```python
df['Quantity']=df['Quantity'].replace('five','5')
df
```

```python
df["Quantity"]=pd.to_numeric(df["Quantity"].fillna(df['Quantity'].mode()[0]))
df
```

```python
df.dtypes
```

```python
df.rename(columns={'TotalAmount':'Sales'},inplace=True)
df
```

```python
df['OrderDate']=pd.to_datetime(df['OrderDate'])
df
```

```python
df['UnitPrice']=df['UnitPrice'].replace('unknown','0')
df
```

```python
df['UnitPrice']=pd.to_numeric(df['UnitPrice'])
df
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
data=df.groupby('Region')['Sales'].sum().reset_index()
data
```

```python
sns.barplot(x='Region',y='Sales',data=data,hue='Region')
plt.show()
```

```python
data1=df.groupby('Product')['Sales'].sum().reset_index()
data1
```

```python
sns.barplot(x='Product',y='Sales',data=data1,hue='Product')
plt.show()
```

```python
data3=df.groupby('Product')['UnitPrice'].mean().reset_index()
data3
```

```python
sns.barplot(x='Product',y='UnitPrice',data=data3,hue='Product')
plt.show()
```

```python
data4=df.groupby('SalesPerson')['Quantity'].sum().reset_index()
data4
```

```python
sns.barplot(x='SalesPerson',y='Quantity',data=data4,hue='SalesPerson')
plt.show()
```

```python
data5=df.groupby(df['OrderDate'].dt.month)['Sales'].sum().reset_index()
data5
```

```python
sns.barplot(x='OrderDate',y='Sales',data=data5,hue='OrderDate')
plt.show()
```
